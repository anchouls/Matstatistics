```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.95
length = []
for n in range(1, 20):
    a = []
    for _ in range(1000):
        x = np.random.normal(0, 1, n)
        sm = np.sum(np.square(x))
        l = sm / sps.chi2.ppf((1+alpha)/2, n)
        r = sm / sps.chi2.ppf((1-alpha)/2, n)
        a.append(r - l)
    length.append(np.mean(a))

plt.figure(figsize=(15,8))
plt.title('$\sum X_k^2 / n $')
plt.plot(np.arange(1,20), np.array(length))
plt.xlim((1,20))
plt.show()
```

```python
alpha = 0.95
length = []
for n in range(1, 50):
    a = []
    for _ in range(1000):
        x = np.random.normal(0, 1, n)
        x_mean = x.mean()
        l = n * (x_mean ** 2) / (sps.norm.ppf((3+alpha)/4))**2
        r = n * (x_mean ** 2) / (sps.norm.ppf((3-alpha)/4))**2
        a.append(r - l)
    length.append(np.mean(a))

plt.figure(figsize=(15,8))
plt.title('$(\sum X_k / n)^2$')
plt.plot(np.arange(1,50), np.array(length))
plt.xlim((1,50))
plt.show()
```
